```python
import pandas as pd
dataset1 = pd.read_excel('./mushrooms.xlsx',dtype=str)
dataset1.head(5)
```

# ID3 w/o Pruning

```python
from id3 import Id3Estimator, export_graphviz, export_text
from sklearn.model_selection import cross_val_score, train_test_split
import pydot
import numpy as np

feature_names = ["cap-shape","cap-surface","cap-color",
                 "bruises","odor","gill-attachment",
                 "gill-spacing","gill-size","gill-color",
                 "stalk-shape","stalk-root","stalk-surface-above-ring",
                 "stalk-surface-below-ring","stalk-color-above-ring","stalk-color-below-ring",
                 "veil-type","veil-color","ring-number",
                 "ring-type","spore-print-color","population",
                 "habitat"]

X = dataset1.drop(['class'], axis=1).to_numpy().astype('S32')
y = dataset1['class'].to_numpy().astype('S32')

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
print('Train data :',len(X_train))
print('Test data :',len(X_test))

clf = Id3Estimator(prune=False)
clf.fit(X_train, y_train, check_input=True)

export_graphviz(clf.tree_, "out.dot", feature_names)
```

```python
import graphviz
[graph,]= pydot.graph_from_dot_file('out.dot')
graph.write_png('ID3NoPruning.png')
pic = graphviz.Source(graph)
pic
```

```python
print(export_text(clf.tree_, feature_names))
```

```python
predict = clf.predict(X_test)

def getPredictAcc(predict,label):
    total = len(predict)
    accuracy = float(np.count_nonzero(label == predict)/total)*100
    return print('Test accuracy: %f \non %d test data'%(accuracy,total))

getPredictAcc(predict,y_test)
```

# ID3 w/ Pruning

```python
from id3 import Id3Estimator, export_graphviz, export_text
from sklearn.model_selection import cross_val_score, train_test_split
import pydot
import numpy as np

feature_names = ["cap-shape","cap-surface","cap-color",
                 "bruises","odor","gill-attachment",
                 "gill-spacing","gill-size","gill-color",
                 "stalk-shape","stalk-root","stalk-surface-above-ring",
                 "stalk-surface-below-ring","stalk-color-above-ring","stalk-color-below-ring",
                 "veil-type","veil-color","ring-number",
                 "ring-type","spore-print-color","population",
                 "habitat"]

X = dataset1.drop(['class'], axis=1).to_numpy().astype('S32')
y = dataset1['class'].to_numpy().astype('S32')

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
print('Train data :',len(X_train))
print('Test data :',len(X_test))

clf = Id3Estimator(prune=True)
clf.fit(X_train, y_train, check_input=True)

export_graphviz(clf.tree_, "out.dot", feature_names)
```
